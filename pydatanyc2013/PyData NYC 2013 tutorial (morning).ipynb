# Example 4: What kinds of noise complaints do people have?

```python
orig_data['Complaint Type'].str.contains("Noise")
```

```python
noise_complaints = orig_data[orig_data['Complaint Type'].str.contains("Noise")]
```

```python
noise_complaints['Complaint Type'].value_counts()
```
